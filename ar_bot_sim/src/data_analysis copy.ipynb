# Training

This section contains code that will train a number of models. 

```python
import gymnasium as gym

import numpy as np

from ar_bot_gym.ar_bot_gym import ARBotGym
from ar_bot_pybullet.ar_bot_pybullet import ARBotPybullet
from stable_baselines_models.train_arbot import TrainARBot

from stable_baselines3 import PPO

NUM_TRIALS = 3
NUM_EPISODES = 1000
```

## Boxed Action Space

```python
actions = gym.spaces.box.Box(
            low=np.array([-0.5, -0.5]),
            high=np.array([0.5, 0.5]),
        )
```

```python
ppo_boxed = TrainARBot(ARBotPybullet, ARBotGym, actions, PPO)
ppo_boxed_total_sum_reward_tracker, ppo_boxed_total_timestep_tracker = ppo_boxed.train(NUM_TRIALS, NUM_EPISODES, "stable_baselines_models/trained_models/ppo_boxed/", "stable_baselines_models/training_data/ppo_boxed.npy")
```

## Discrete Action Space

```python
actions = gym.spaces.Discrete(4)

action_mapping = {
    0: (0.0, 0.5),
    1: (0.5, 0.0),
    2: (0.0, -0.5),
    3: (-0.5, 0.0)
}
```

```python
ppo_discrete = TrainARBot(ARBotPybullet, ARBotGym, actions, PPO, action_mapping)
ppo_discrete_total_sum_reward_tracker, ppo_discrete_total_timestep_tracker = ppo_discrete.train(NUM_TRIALS, NUM_EPISODES, "stable_baselines_models/trained_models/ppo_discrete/", "stable_baselines_models/training_data/ppo_discrete.npy", obstacle=True)
```

## MultiDiscrete Action Space

```python
actions = gym.spaces.MultiDiscrete([3, 3])

action_mapping = {
    0: -0.5,
    1: 0,
    2: 0.5
}
```

```python
ppo_multi_discrete = TrainARBot(ARBotPybullet, ARBotGym, actions, PPO, action_mapping)
ppo_multi_discrete_total_sum_reward_tracker, ppo_multi_discrete_total_timestep_tracker = ppo_multi_discrete.train(NUM_TRIALS, NUM_EPISODES, "stable_baselines_models/trained_models/ppo_multi_discrete/", "stable_baselines_models/training_data/ppo_multi_discrete.npy")
```

# Print Graphs



## Setup
The importing of numpy is redandant if running in the same kernal used for training

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter1d
```

## Parse

Parse the data contained in the npy files created during training above. 

This step can be safely skipped if running in the same kernal that training was completed in.

Note that the file locations will needed to be tweaked if using a different training_data directory.

```python
with open("stable_baselines_models/training_data/good_data/ppo_boxed.npy", "rb") as ppo_data:
    ppo_total_sum_reward_tracker = np.load(ppo_data)
    ppo_total_timestep_tracker = np.load(ppo_data)
```

## Create Graphs

The following references were used to create graphs:     
    https://stackoverflow.com/questions/61799259/plot-smoothing-matplotlib-and-seaborn
    https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.gaussian_filter1d.html

```python
smoothed_ppo_total_timestep_tracker = gaussian_filter1d(ppo_total_timestep_tracker, sigma=10)

plt.plot(smoothed_ppo_total_timestep_tracker, label="PPO")

plt.legend()
plt.ylabel("Number of Timesteps")
plt.xlabel("Episode Number")
plt.title("Number of Timesteps Per Episode Episodes")

plt.show()
```

```python
smoothed_ppo_total_sum_reward_tracker = gaussian_filter1d(ppo_total_sum_reward_tracker, sigma=10)

plt.plot(smoothed_ppo_total_sum_reward_tracker, label="PPO")

plt.legend()
plt.ylabel("Sum of Rewards")
plt.xlabel("Episode Number")
plt.title("Sum Of Rewards Across Episodes")

plt.show()
```
